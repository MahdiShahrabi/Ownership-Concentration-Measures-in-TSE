```python
import jdatetime as jd

def find_jdate(x):
    year = int(x/10000)
    month = int((x%10000)/100)
    day = int((x%100))
    out = jd.date(year,month,day)
    return(out)
    

gb_inst['new_date'] = 10000*gb_inst.year + 100*gb_inst.month+gb_inst.month
gb_inst['jdate']= gb_inst['new_date'].apply(lambda x: find_jdate(x))

gb_inst.drop(columns=['year','month','day','new_date'],inplace=True)
gb_inst['jdate_str'] = gb_inst['jdate'].apply(lambda x: str(x))

gb_inst.head()
```
