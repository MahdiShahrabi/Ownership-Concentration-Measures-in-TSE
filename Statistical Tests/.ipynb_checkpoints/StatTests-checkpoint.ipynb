# **Loading Data**

```python
## Loading DATA
os.chdir(r"C:\Users\Mahdi\OneDrive\Master Thesis\Data")
df = pd.read_csv("data_owenership97_blnc98.csv",index_col=0)
OC_list = ['Largest_Owner','First_Second','First_Sumtwofour','Sumfive',
           'Gini','Herfindhal','SSCL','SSCO','SSDL','SSDO','BZCL','BZCO','BZDL']
```

## Wilcoxon Test

```python
out = pd.DataFrame()
out_list = list()

for i in range(len(OC_list)):
    x = df[OC_list[i]]
    
    for j in range(i+1,len(OC_list)):
        y = df[OC_list[j]]

        w,p = stats.wilcoxon(x,y)
        out.loc[OC_list[j],OC_list[i]] = p
        
        out_list.append([OC_list[j],OC_list[i],w,p])
        
out

os.chdir(r"C:\Users\Mahdi\OneDrive\Master Thesis\Data")
out.to_csv('Wilcoxon_p.csv')
```
