## Wilcoxon Test

```python
out = pd.DataFrame()
out_list = list()

for i in range(len(OC_list)):
    x = df[OC_list[i]]
    
    for j in range(len(OC_list)):
        y = df[OC_list[j]]
        
        zipped = [ (a,b) for a,b in zip(x,y) if a is not np.nan and b is not np.nan]
        x =  [a for a,b in zipped]
        y =  [b for a,b in zipped]

        w,p = stats.wilcoxon(x=x,y=y)
        
        if p<0.05 and p>0.01:
            out.loc[OC_list[j],OC_list[i]] = str(np.round(w,3))+'*'
        elif p<0.01:
            out.loc[OC_list[j],OC_list[i]] = str(np.round(w,3))+'**'
        else:
            out.loc[OC_list[j],OC_list[i]] = str(np.round(w,3))
        
        out_list.append([OC_list[j],OC_list[i],w,p])
        
out

os.chdir(r"C:\Users\Mahdi\OneDrive\Master Thesis\Data")
out.to_csv('Wilcoxon_p.csv')
```

## Spearman Correlation Rank Test

```python
out = pd.DataFrame()
out_list = list()

for i in range(len(OC_list)):
    x = df[OC_list[i]]
    
    for j in range(len(OC_list)):
        y = df[OC_list[j]]
        
        zipped = [ (a,b) for a,b in zip(x,y) if a is not np.nan and b is not np.nan]
        x =  [a for a,b in zipped]
        y =  [b for a,b in zipped]

        rho,p = stats.spearmanr(x,y,nan_policy='omit')
        if p<0.05 and p>0.01:
            out.loc[OC_list[j],OC_list[i]] = str(np.round(rho,4))+'*'
        elif p<0.01:
            out.loc[OC_list[j],OC_list[i]] = str(np.round(rho,4))+'**'
        else:
            out.loc[OC_list[j],OC_list[i]] = str(np.round(rho,4))
        
        out_list.append([OC_list[j],OC_list[i],w,p])
        
out

os.chdir(r"C:\Users\Mahdi\OneDrive\Master Thesis\Data")
out.to_csv('Spearman_p.csv')
```
